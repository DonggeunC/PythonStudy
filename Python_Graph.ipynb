```python
import numpy as np                        #모듈추가
import matplotlib.pyplot as plt             
%matplotlib inline                           

x = np.linspace(-4,4, 100)                 #데이터 생성
                                             
y1 = 2 ** x                               
y2 = 3 ** x                               
y3 = 0.5 ** x                               

print('x = {}'.format(x))                 #데이터 로그 확인
print('y1 = {}'.format(y1))                 
print('y2 = {}'.format(y2))               
print('y3 = {}'.format(y3))                 

plt.figure(figsize=(5,5))                                                #그래프 출력
plt.plot(x,y1,'black',linewidth = 3, label = '$y=2^x$')                    
plt.plot(x,y2,'cornflowerblue',linewidth = 3, label = '$y=3^x$')
plt.plot(x,y3,'gray',linewidth = 3, label = '$y=0.5^x$')

plt.ylim(-2,6)
plt.xlim(-4.4)
plt.grid(True)
plt.legend(loc = 'lower right')
#plt.show()
plt.savefig('graph.png')                                                   
```

# 3차 함수 ; f(x) = (x-2)x(x+2)

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    y = (x - 2) * x * (x + 2)
    return y

print('x = 1 -> y = {}'.format(f(1)))
print('x = 2 -> y = {}'.format(f(2)))
print('x = 3 -> y = {}'.format(f(3)))

x = np.array([1, 2, 3])   #list -> np.ndarray
x = np.arange(-3.3, 5, 0.5)  #np.arange => 범위지정
x = np.linspace(-3, 3, 50)   #start, finish, samples

print (f'x[] = {x}, type(x) = {type(x)}')
print(f(x))

plt.plot(x,f(x),'cornflowerblue',linewidth = 3, label = 'y=(x-2)*x*(x+2)')
plt.xlim(-4,4)
plt.ylim(-20,20)
plt.grid(True)
plt.legend(loc = 'lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, w):
    y = (x - w) * x * (x + 2)
    return y

print('x = 1 -> y = {}'.format(f(1,5)))
print('x = 2 -> y = {}'.format(f(2,5)))
print('x = 3 -> y = {}'.format(f(3,5)))

x = np.array([1, 2, 3])   #list -> np.ndarray
x = np.arange(-3.3, 5, 0.5)  #np.arange => 범위지정
x = np.linspace(-4, 4, 50)   #start, finish, samples
#w = np.linspace(-5, 5, 50)

print (f'x[] = {x}, type(x) = {type(x)}')
#print(f(x,w))

plt.plot(x,f(x,w=0),'cornflowerblue',linewidth = 3, label = '$y=(x-0)*x*(x+2)$')
plt.plot(x,f(x,w=4),'red',linewidth = 3, label = '$y=(x-4)*x*(x+2)$')
plt.plot(x,f(x,w=2),'black',linewidth = 3, label = '$y=(x-2)*x*(x+2)$')
plt.plot(x,f(x,w=np.arange(-7,7,40)),'indigo', linewidth = 3, label = '$y=(x-w)*x*(x+2)$')
plt.plot(x,f(x,w=10),'chartreuse',linewidth = 3, label = '$y=(x-10)*x*(x+2)$')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title('$f(x , w)$')
plt.xlim(-4,4)
plt.ylim(-40,40)
plt.grid(True)
plt.legend(loc = 'lower right')
plt.show()
```

```python
import numpy as np

def f3(x0, x1):
    r = 2 * x0 **2 + x1 **2
    ans = r * np.exp(-r)
    return ans

xn = 10
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)

y = np.zeros((len(x0),len(x1)))

print('x0 = {}'.format(x0))
print('x1 = {}'.format(x1))
print('y = {}'.format(y))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0] = f3(x0[i0],x1[i1])
y = np.round(y,1)        
print('y = {}'.format(y))

plt.figure(figsize = (3.5,3))
plt.gray()
#plt.pcolor(y)
plt.contour(y)

plt.colorbar()
plt.savefig('gauss.png')
```

# Quiz. 구구단 3단, 5단,7단, 9단을 Plot(1개), SubPlot(4분할)로 그래프 표현

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x):
        num = np.linspace(1,9,9)
        return x*num

plt.figure(figsize=(4,4))
x = np.linspace(1,9,9)

y1 = f(3)
y2 = f(5)
y3 = f(7)
y4 = f(9)

plt.plot(x,y1,'red',label = '$y = 3 * x$')
plt.plot(x,y2,'black',label = '$y = 5 * x$')
plt.plot(x,y3,'cornflowerblue',label = '$y = 7 * x$')
plt.plot(x,y4,'gold',label = '$y = 9 * x$')

plt.xlim(1,9)
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.legend(loc = 'upper left')
plt.grid(True)

plt.figure(figsize=(6,6))
plt.subplots_adjust(wspace = 0.5,hspace = 0.5)
for i in range(1,5):
    y = f(2*i+1)
    plt.subplot(2,2,i)
    plt.title(str((2*i+1))+'Dan')
    plt.xlim(1,9)
    plt.ylim(0,100)
    plt.plot(x,y,'k')
    plt.grid(True)
```
