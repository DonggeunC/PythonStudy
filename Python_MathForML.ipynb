# 미분

f(x) = x^2 .. x에 대한 식(1)  
dx_f(x) = 2*x .. 식(1)에 대해서 x에 대한 1차 미분  
  
f(x,w) = x^2+w^2 .. x에 대한 식(2)  
dx_f(x,w) = 2*x    .. 식(2)에 대해서 x에 대한 1차미분  
dw_f(x,w) = 2*w    .. 식(2)에 대해서 w에 대한 1차미분  

```python
def f(w0,w1):
    return w0**2 + 2*w0*w1+3        # x^2 +2xy +3

def df_dw0(w0,w1):
    return 2*w0 + 2*w1              # 2x + 2y

def df_dw1(w0,w1):
    return 2*w0+0*w1                     # 2x
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(w0,w1):
    return w0**2 + 2*w0*w1+3 

w_range = 4
dw = 0.25
w0 = np.arange(-w_range, w_range+dw, dw)
w1 = np.arange(-w_range, w_range+dw, dw)

#y = np.zeros((len(w0),len(w1)))  
ww0,ww1 = np.meshgrid(w0,w1)

y = f(ww0,ww1)

plt.figure(figsize = (4,4))

ax=plt.subplot(1,1,1,projection = '3d')
ax.plot_surface(ww0,ww1,y,rstride = 1, cstride = 1, alpha = 0.5, color = 'black', edgecolor = 'moccasin')

plt.xlabel('$W0$',fontsize = 8)
plt.ylabel('$W1$',fontsize = 8)

ax.view_init(20,120)
```

```python
y1 =f(w0,-1)
dy1 = df_dw0(w0,-1)

plt.figure(figsize =(4,4))
plt.plot(w0,y1,'black',linewidth=3)
plt.plot(w0,dy1,'cornflowerblue',linewidth=3)
plt.xlim(-5,5)
plt.ylim(-10,10)
plt.xlabel('$w0$')
plt.ylabel('$Y$')
plt.title('$W1 = -1$')
plt.grid(True)
```

```python
ti = [-1,1,-1,1]

plt.figure(figsize=(6,6))
plt.subplots_adjust(wspace = 0.5,hspace = 0.5)

for i in range(4):
    plt.subplot(2,2,i+1)
    if i<2:
        plt.plot(w0,f(w0,ti[i]),'salmon',linewidth = 2, label = '$w_1= {}$'.format(ti[i]))
        plt.plot(w0,df_dw0(ti[i],w1),'cornflowerblue',linewidth = 2,label = '$dw_0$')
        plt.title('$w_1 = {}$'.format(ti[i]))
        plt.xlabel('$w_0$')
    elif i>=2:
        plt.plot(w1,f(ti[i],w1),'salmon',linewidth = 2, label = '$w_0= {}$'.format(ti[i]))
        plt.plot(w1,df_dw1(ti[i],w1),'cornflowerblue',linewidth = 2,label = '$dw_1$')
        plt.title('$w_0 = {}$'.format(ti[i]))
        plt.xlabel('$w_1$')
    plt.xlim(-2,2)
    plt.ylim(-10,10)
    plt.ylabel('$y$')
    plt.legend(loc = 'lower right')
    plt.grid(True)

```

```python
wn = w0.shape[0]
ww0,ww1 = np.meshgrid(w0,w1)
ff= np.zeros((len(w0),len(w1)))
dff_dw0 = np.zeros((len(w0),len(w1)))
dff_dw1 = np.zeros((len(w0),len(w1)))

for i0 in range(wn):
    for i1 in range(wn):
        ff[i1,i0] = f(w0[i0],w1[i1])
        dff_dw0[i1,i0] = df_dw0(w0[i0],w1[i1])
        dff_dw1[i1,i0] = df_dw0(w0[i0],w1[i1])
        
plt.figure(figsize = (9,4))
plt.subplots_adjust(wspace = 0.3)
plt.subplot(1,2,1)
cont = plt.contour(ww0,ww1,ff,10,colors = 'k')
cont.clabel(fmt = '%2.0f',fontsize = 8)
plt.xticks(range(-w_range,w_range+1,1))
plt.yticks(range(-w_range,w_range+1,1))
plt.xlim(-w_range-0.5,w_range+.5)
plt.ylim(-w_range-.5,w_range+.5)
plt.xlabel('$w_0$',fontsize = 14)
plt.ylabel('$w_1$',fontsize = 14)

plt.subplot(1,2,2)
plt.quiver(ww0,ww1,dff_dw0,dff_dw1)
plt.xlabel('$w_0$',fontsize = 14)
plt.ylabel('$w_1$',fontsize = 14)
plt.xticks(range(-w_range,w_range+1,1))
plt.yticks(range(-w_range,w_range+1,1))
plt.xlim(-w_range-0.5,w_range+.5)
plt.ylim(-w_range-.5,w_range+.5)
```

```python
#gaussian 함수
def gauss(mu,sigma,a):
    return a*np.exp(-(x-mu)**2/sigma**2)

x=np.linspace(-4,4,100)
plt.figure(figsize = (4,4))
plt.plot(x,gauss(0,1,1),'black',linewidth =3 )
plt.plot(x,gauss(2,3,0.5),'gray',linewidth =3 )
plt.ylim(-.5,1.5)
plt.xlim(-4,4)
plt.grid(True)
plt.show()
```

```python
def gauss(x,mu,sigma):
    N,D = x.shape
    c1 = 1/(2*np.pi)**(D/2)
    c2 = 1/(np.linalg.det(sigma)**(1/2))
    inv_sigma = np.linalg.inv(sigma)
    c3 =x-mu
    c4 = np.dot(c3,inv_sigma)
    c5 = np.zeros(N)
    for d in range(D):
        c5 = c5+c4[:,d]*c3[:,d]
    p = c1*c2*np.exp(-c5/2)
    return p
```

```python
x=np.array([[1,2],[2,1],[3,4]])
mu = np.array([1,2])
sigma = np.array([[1,0],[0,1]])
print(gauss(x,mu,sigma))
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_range0 = [-3,3]
x_range1 = [-3,3]

def show_contour_gauss(mu,sig):
    xn = 40
    x0 = np.linspace(x_range0[0],x_range0[1],xn)
    x1 = np.linspace(x_range1[0],x_range1[1],xn)
    xx0,xx1 = np.meshgrid(x0,x1)
    
    x = np.c_[np.reshape(xx0,(xn*xn,1)),np.reshape(xx1,(xn*xn,1))]
    f = gauss(x,mu,sig)
    f=f.reshape(xn,xn)
    f=f.T
    cont = plt.contour(xx0,xx1,f,15,colors = 'k')
    plt.grid(True)
    
def show3d_gauss(ax,mu,sig):
    xn = 40
    x0 = np.linspace(x_range0[0],x_range0[1],xn)
    x1 = np.linspace(x_range1[0],x_range1[1],xn)
    xx0,xx1 = np.meshgrid(x0,x1)
    
    x = np.c_[np.reshape(xx0,(xn*xn,1)),np.reshape(xx1,(xn*xn,1))]
    f = gauss(x,mu,sig)
    f=f.reshape(xn,xn)
    f=f.T
    ax.plot_surface(xx0,xx1,f,rstride =2,cstride = 2, alpha = 0.3, color = 'blue',edgecolor = 'black')
    
mu = np.array([1,0.5])
sigma = np.array([[2,1],[1,1]])
Fig = plt.figure(1,figsize=(7,3))
Fig.add_subplot(1,2,1)
show_contour_gauss(mu,sigma)
plt.xlim(x_range0)
plt.ylim(x_range1)
plt.xlabel('$x_0$',fontsize = 14)
plt.ylabel('$x_1$',fontsize = 14)
Ax = Fig.add_subplot(1,2,2,projection = '3d')
show3d_gauss(Ax,mu,sigma)
Ax.set_zticks([0.05,0.10])
Ax.set_xlabel('$x_0$',fontsize = 14)
Ax.set_ylabel('$x_1$',fontsize = 14)
Ax.view_init(40,-100)
plt.show()
```

```python
#softmax 함수

def softmax(x0,x1,x2):
    u = np.exp(x0)+np.exp(x1)+np.exp(x2)
    return np.exp(x0) / u, np.exp(x1) /u, np.exp(x2) /u

y= softmax(2,1,-1)
print(np.round(y,2))
print(np.sum(y))
```

```python
xn = 20
x0 = np.linspace(-4,4,xn)
x1 = np.linspace(-4,4,xn)

y = np.zeros((xn,xn,3))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0,:] = softmax(x0[i0],x1[i1],1)

xx0,xx1 = np.meshgrid(x0,x1)
plt.figure(figsize=(8,3))
for i in range(2):
    ax = plt.subplot(1,2,i+1,projection = '3d')
    ax.plot_surface(xx0,xx1,y[:,:,i],rstride = 1, cstride = 1, alpha = 0.3, color ='blue',edgecolor = 'black')
    ax.set_xlabel('$x_0$',fontsize=14)
    ax.set_ylabel('$x_1$',fontsize=14)
    ax.view_init(40,-125)
    
plt.show()
```

```python
import cv2

# 얼굴 검출기 초기화
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 웃음 검출기 초기화
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_smile.xml')

# 이미지 불러오기
img = cv2.imread('image.jpg')

# 이미지를 흑백으로 변환
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 얼굴 검출
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# 검출된 얼굴에 사각형 표시
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
    # 얼굴 ROI
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]

    # 웃음 검출
    smiles = smile_cascade.detectMultiScale(roi_gray)
    for (sx, sy, sw, sh) in smiles:
        if sw >90 and sh >85:
            cv2.rectangle(roi_color, (sx, sy), (sx+sw, sy+sh), (255, 0, 0), 2)

# 결과 출력
cv2.imshow('img', img)
cv2.waitKey()
```
